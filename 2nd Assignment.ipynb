```python
#### SECOND ASSIGNMENT ####
#Course: Python
#Week: 2
#ID: 318369329
#Name: Geffen Horowitz

import pandas as pd
import numpy as np

from google.colab import files

# create variables
participant_id = np.arange(1, 101)
age = np.random.uniform(15, 41, 100)
age = np.round(age)
gender = np.random.choice(["male", "female"], size=100, replace=True)
iq = np.random.normal(loc=100, scale=15, size=100)
depression = np.random.binomial(n=1, p=0.15, size=100)
anxiety = np.random.uniform(0, 101, 100)

#create df
df = pd.DataFrame({
    "participant_id" : participant_id,
    "age" : age,
    "iq" : iq,
    "gender" : gender,
    "depression" : depression,
    "anxiety" : anxiety
})

#save to CSV
df.to_csv("df_2nd_assignment.csv", index=False, encoding="utf-8-sig")

#describe
df["age"].describe()
df.describe()

#plot
ax = df["age"].hist(bins=10)
ax.set_xlabel("Age")


```
